```python
import re
def clean_column_name(col):
    cleaned = re.sub(r'[^a-zA-Z\s]', '', col).strip().lower().replace(" ", "_")
    first_three_words = "_".join(cleaned.split("_")[:4])
    return first_three_words
```

```python
column_labels = {clean_column_name(col): col for col in df.columns}
column_labels
```
